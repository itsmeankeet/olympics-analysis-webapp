```python
def fetch_medal_tally(df, year, country):
    medal_df = df.drop_duplicates(subset=['Team', 'NOC', 'Games', 'Year', 'City', 'Sport', 'Event', 'Medal'])
    flag = 0
    if(year == 'Overall' and country == 'Overall'):
        temp_df = medal_df
    if(year == 'Overall' and country != 'Overall'):
        flag = 1
        temp_df = medal_df[medal_df['region'] == country]
    if(year != 'Overall' and country == 'Overall'):
        temp_df = medal_df[medal_df['Year'] == int(year)]
    if(year != 'Overall' and country != 'Overall'):
        temp_df = medal_df[(medal_df['Year'] == int(year)) & (medal_df['region'] == country)]
    if(flag == 1): 
        temp_df = temp_df.groupby('Year').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Year', ascending=True).reset_index()
    else:
        temp_df = temp_df.groupby('region').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending=False).reset_index()
    temp_df['total'] = temp_df['Gold'] + temp_df['Silver'] + temp_df['Bronze']
    return temp_df

fetch_medal_tally(df, year='1900', country='India')

```

```python
# This code counts the number of unique regions (countries) participating in each Olympic year:
# 1. df.drop_duplicates(['Year','region']) - Removes duplicate Year-region combinations to count each country only once per year
# 2. ['Year'] - Selects just the Year column after deduplication
# 3. value_counts() - Counts how many times each Year appears, which gives us the number of unique regions per year
# 4. reset_index() - Makes Year a column instead of index
# 5. sort_values('Year') - Sorts the results chronologically by Year
# 1. Removes duplicate rows that have the same Year and region using drop_duplicates()
# 2. Selects just the Year column 
# 3. Counts how many unique regions participated each year using value_counts()
# 4. Resets the index to make Year a column instead of index
# 5. Sorts the results by Year in ascending order
# 
# The output shows how many countries participated in each Olympic year
nation_over_time = df.drop_duplicates(['Year','region'])['Year'].value_counts().reset_index().sort_values('Year')
```

```python
def most_successful(df, sport):
    """Returns DataFrame of most successful athletes by medal count for a given sport"""
    # Filter for valid medals and sport
    temp_df = df[df['Medal'].notna()]
    if sport != 'Overall':
        temp_df = temp_df[temp_df['Sport'] == sport]
    
    # Group and count medals
    medals = temp_df.groupby(['Name', 'region', 'Sport'])['Medal'].count().reset_index()
    medals = medals.sort_values('Medal', ascending=False)
    medals.columns = ['Name', 'Region', 'Sport', 'Medals']
    return medals
```
